```python
df[df['Outcome'] == 'Successful']
```

# countries with the most successful missions

```python
most_success = df[df['Outcome'].isin(['Successful','Operational'])]
most_success['Operator'].value_counts().head(5).plot(kind='bar', title='most successful countries')
```

# most failed mission

```python
most_failed = df[df['Outcome'].isin(['Launch failure','Partial failure','Spacecraft failure'])]
most_failed['Operator'].value_counts().head(6).plot(kind='bar', title='Most failed')
```

```python
succ_type = df[df['Outcome'].isin(['Successful','Operational'])]
succ_type['Mission Type'].value_counts()
```

```python
fail_type = df[df['Outcome'].isin(['Launch failure','Partial failure','Spacecraft failure'])]
fail_type['Mission Type'].value_counts()
```

# Top 10 Spacecraft

```python
# Top 10 Spacecraft
top_spacecraft = df['Spacecraft'].value_counts().head(10)
print("Top 10 Spacecraft:")
for idx, (spacecraft, count) in enumerate(top_spacecraft.items(), 1):
    print(f"Spacecraft {idx}: {spacecraft} - Count {count}")
```

# Top 5 Carrier Rockets

```python
# Top 5 Carrier Rockets
top_carrier_rockets = df['Carrier Rocket'].value_counts().head(5)
print("\nTop 5 Carrier Rockets:")
for idx, (rocket, count) in enumerate(top_carrier_rockets.items(), 1):
    print(f"Carrier Rocket {idx}: {rocket} - Count {count}")
```

# Deductions-


**There have been a total of 161 lunar missions recorded in the dataset.**

**The United States (US) and the Union of Soviet Socialist Republics (USSR) are the leading countries in lunar missions, indicating their significant contributions to lunar exploration.**

**The first successful moon mission recorded in the dataset was 'Luna 2 (E-1A No.2)', launched on 12-Sep-59 by the Soviet Union. This mission marked a historic achievement in lunar exploration.**

**The most successful lunar missions were conducted by the United States (US), USSR, and China. The US leads with 45 successful missions, followed by the USSR and China.**

**The USSR holds the record for the highest number of failed lunar missions, with a total of 36 failures. The United States follows closely with 31 failed missions.**

**The most common type of mission that resulted in failure is identified as "Lander Missions." This suggests that landing missions have historically presented challenges and complexities.**
