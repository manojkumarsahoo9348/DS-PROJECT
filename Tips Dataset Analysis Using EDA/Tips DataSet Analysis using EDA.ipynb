# 2. Gender spending paterns

**By analyzing total bills based on gender, you can see if there are any differences in spending between males and females. Are there any noticeable patterns or variations?**

```python
# Assuming your dataset is stored in a DataFrame named 'tips_df'
gender_spending_stats = tips_df.groupby('sex').agg({
    'total_bill': ['mean', 'sum'],
    'tip': ['mean', 'sum'],
    'size': 'count'
}).reset_index()
```

```python
# Create visualizations
plt.figure(figsize=(12, 6))

# Barplot for Average Total Bill
plt.subplot(1, 2, 1)
sns.barplot(data=gender_spending_stats, x='sex', y=('total_bill', 'mean'))
plt.title('Average Total Bill by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Total Bill')

# Barplot for Total Number of Visits
plt.subplot(1, 2, 2)
sns.barplot(data=gender_spending_stats, x='sex', y=('size', 'count'))
plt.title('Total Number of Visits by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Number of Visits')

plt.tight_layout()
plt.show()
```

# 8. Find the number of male and female per day

```python
# Group the data by "day" and "sex" and count the occurrences
gender_per_day = tips_df.groupby(['day', 'sex']).size().unstack()

# Fill any missing values with 0 (in case there are no occurrences of a particular gender on a day)
gender_per_day = gender_per_day.fillna(0)

# Display the result
print(gender_per_day)
```

```python
# Create a bar plot to visualize the number of males and females coming each day
plt.figure(figsize=(10, 6))
sns.barplot(data=gender_per_day, x=gender_per_day.index, y='Male', label='Male', color='blue')
sns.barplot(data=gender_per_day, x=gender_per_day.index, y='Female', label='Female', color='red')
plt.title('Number of Males and Females Coming Each Day')
plt.xlabel('Day')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
sns.countplot(x='day',data=tips_df,hue='sex')
plt.show()
```

# 9.total_bill for genders per day

```python
# Group the data by "day" and "sex" and sum the total bill for each group
total_bill_per_gender_per_day = tips_df.groupby(['day', 'sex'])['total_bill'].sum().unstack()

# Fill any missing values with 0 (in case there are no occurrences of a particular gender on a day)
total_bill_per_gender_per_day = total_bill_per_gender_per_day.fillna(0)

# Display the result
print(total_bill_per_gender_per_day)
```

```python
# Create a bar plot to visualize the total bill for different genders per day
plt.figure(figsize=(10, 6))
sns.barplot(data=tips_df, x='day', y='total_bill', hue='sex', estimator=sum, ci=None)
plt.title('Total Bill for Different Genders Per Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.legend(title='Gender', title_fontsize='12')
plt.show()
```

# 10. Party Size and Total Bill: 
   **Investigate whether the size of the party correlates with the total bill. Do larger groups tend to spend more?**

```python
# Assuming your dataset is stored in a DataFrame named 'df'
party_size_stats = tips_df.groupby('size').agg({
    'total_bill': ['mean', 'sum'],
    'tip': ['mean', 'sum']
}).reset_index()
```

```python
# Create visualizations
plt.figure(figsize=(12, 6))

# Scatterplot for Average Total Bill by Party Size
plt.subplot(1, 2, 1)
sns.scatterplot(data=party_size_stats, x='size', y=('total_bill', 'mean'))
plt.title('Average Total Bill by Party Size')
plt.xlabel('Party Size')
plt.ylabel('Average Total Bill')

# Scatterplot for Total Bill Sum by Party Size
plt.subplot(1, 2, 2)
sns.scatterplot(data=party_size_stats, x='size', y=('total_bill', 'sum'))
plt.title('Total Bill Sum by Party Size')
plt.xlabel('Party Size')
plt.ylabel('Total Bill Sum')

plt.tight_layout()
plt.show()
```
